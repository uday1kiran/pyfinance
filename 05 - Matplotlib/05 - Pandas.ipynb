```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
import random
```

```python
s1 = pd.Series([random.uniform(0, 1) for i in range(30)])
```

```python
fig1, ax1 = plt.subplots()
```

```python
s1.plot(ax=ax1)
```

```python
s2 = pd.Series([random.uniform(0, 1) for i in range(30)])
```

```python
s2.plot(ax=ax1)
```

```python
df = pd.DataFrame({'A': [random.uniform(0, 1) for i in range(100)], 'B': [random.uniform(0, 1) for i in range(100)]})
```

```python
fig2, ax2 = plt.subplots()
```

```python
df.plot(ax=ax2)
```

```python
ax2.set_title('My data')
```

```python
df2 = pd.DataFrame({'A': [random.uniform(0, 1) for i in range(100)], 'B': [random.uniform(0, 10) for i in range(100)]})
```

```python
fig3, ax3 = plt.subplots()
```

```python
df2.plot(ax=ax3)
```

```python
fig4, ax4 = plt.subplots()
```

```python
df2['A'].plot(ax=ax4)
df2['B'].plot(ax=ax4, secondary_y=True)
```
