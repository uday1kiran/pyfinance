```python
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
plt.plot([i for i in range(10)], color='g', linestyle='--', label='A')
plt.plot([i + 1 for i in range(10)], color='r', label='B')
plt.legend()
plt.title("My awesome title")
plt.xlabel("X label")
plt.ylabel("Y label")
```

```python
fig, ax = plt.subplots()
ax.plot([i for i in range(10)], color='g', linestyle='--', label='A')
ax.plot([i + 1 for i in range(10)], color='r', label='B')
ax.legend()
ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_title("This is my title")
```
