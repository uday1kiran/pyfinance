```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
import random
```

```python
s = pd.Series([random.randrange(0, 10) for i in range(10)], index=list('abcdefghij'))
```

```python
s
```

```python
fig1, ax1 = plt.subplots()
```

```python
s.plot.bar(ax=ax1)
```

```python
fig2, ax2 = plt.subplots(2, 1)
```

```python
s.plot.bar(ax=ax2[0])
```

```python
s.plot.barh(ax=ax2[1])
```
