```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
tickers = ["FB", "NVDA", "AMZN", "HP", "AAPL"]

df_list = []
for ticker in tickers:
    df = pd.read_csv("ticker-" + ticker.lower() + ".csv", index_col=0, parse_dates=True)
    df_list.append(df['Adj Close'])

data = pd.concat(df_list, axis=1)
data.columns = tickers
```

```python
data = data/data.iloc[0]
```

```python
data
```

```python
portfolio = np.array([.2, .15, .3, .25, .1])
```

```python
portfolio*10000
```

```python
data*portfolio*10000
```

```python
np.sum(data*portfolio*10000, axis=1)
```

```python
fig1, ax1 = plt.subplots()
np.sum(data*portfolio*10000, axis=1).plot(ax=ax1)
```
