```python
import numpy as np
```

```python
def roll_dice():
    return np.sum(np.random.randint(1, 7, 2))
```

```python
def monte_carolo_simulation(runs=1000):
    result = np.zeros(2)
    for _ in range(runs):
        if roll_dice() == 7:
            result[0] += 1
        else:
            result[1] += 1
    return result/runs
```

```python
monte_carolo_simulation()
```

```python
d1 = np.arange(1, 7)
d2 = np.arange(1, 7)
```

```python
results = np.zeros(1000)

for i in range(1000):
    results[i] = monte_carolo_simulation()[0]
```
