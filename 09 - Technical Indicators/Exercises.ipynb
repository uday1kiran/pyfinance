### Step 3
- We will calculate the Relative Strenght Index (RSI)
- It is a leading indicator
- The formula is a bit involved, but we will break it down
- We first need to get the difference of current Close price with previous.
    - This can be done using .diff()
- Create a series **delta** with the diffe of Close.
- Take head of it.

Resources
- https://www.investopedia.com/terms/r/rsi.asp
- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.diff.html

### Step 4
- Calculate **up** and **down** (defined below here)
- **up** is assigned to be **delta** if the value of **delta** is above zero.
    - This can be done by using .clip(lower=0)
- **down** is assigned to be (minus) **-delta** if the value of **delta** is below zero.
    - This can be done by using .clip(upper=0)
- Take head of up and down (comma separated)

Resources
- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.clip.html

### Step 5
- Calculate the Exponential Moving Average of up and down with decay 13 .
- Assign it to each variable **ema_up** and **ema_down**
    - Use **ewm(com=13, adjust=False).mean()**
- Take head of both ema_up and ema_down

Resources
- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html
- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.mean.html

### Step 6
- Calculate the **rs**
- **rs** is **ema_up** divided by **ema_down**
- Take tail on **rs**

### Step 7
- Calculate the **RSI** indicator and add it as a column to the stock time series data
- **RSI** is $100 - (100/(1 + rs))$
- Take tail of your data
