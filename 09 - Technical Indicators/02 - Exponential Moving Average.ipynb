# Exponential Moving Average
- https://www.investopedia.com/terms/m/movingaverage.asp

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
fb = pd.read_csv("ticker-fb.csv", index_col=0, parse_dates=True)
```

```python
fb.head()
```

```python
fb['MA50'] = fb['Close'].rolling(50).mean()
```

```python
fb.tail()
```

```python
fb['EMA50'] = fb['Close'].ewm(span=50, adjust=False).mean()
```

```python
fb.tail()
```

```python
fig, ax = plt.subplots()
fb[['Close', 'MA50', 'EMA50']].loc['2020-01-01':].plot(ax=ax)
```
