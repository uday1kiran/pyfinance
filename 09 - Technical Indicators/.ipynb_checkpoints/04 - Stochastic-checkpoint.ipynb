### Stochastic oscillator
- Lagging indicator
- https://www.investopedia.com/terms/s/stochasticoscillator.asp

**Calculations**
- 14-high: Maximum of last 14 trading days
- 14-low: Minimum of last 14 trading days
- %K: (Last close - 14-low)*100 / (14-high - 14-low)
- %D: Simple Moving Average of %K

**Signal**
- %K > %D

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
ticker = pd.read_csv("ticker-fb.csv", index_col=0, parse_dates=True)
```

```python
high14 = ticker['High'].rolling(14).max()
low14 = ticker['Low'].rolling(14).min()
# %K: (Last close - 14-low)*100 / (14-high - 14-low)
ticker['%K'] = pct_k = (ticker['Close'] - low14)*100 / (high14 - low14)
ticker['%D'] = pct_d = pct_k.rolling(3).mean()
ticker['Signal'] = pct_k > pct_d
```

```python
ticker.tail()
```

```python
fig, ax = plt.subplots()
ticker.loc['2020-01-01':'2020-07-01'][['%K', '%D']].plot(ax=ax, alpha=0.5)
ax.axhline(80, c='r', alpha=0.3)
ax.axhline(20, c='r', alpha=0.3)
ticker.loc['2020-01-01':'2020-07-01']['Close'].plot(ax=ax, secondary_y=True)
```
