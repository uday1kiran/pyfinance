### Moving Average
- https://www.investopedia.com/terms/m/movingaverage.asp

$MAn_i = \frac{C_{i-(n-1)} + C_{i-(n-2)} + \cdots + C_{i-1} + C_i}{n}$
- $MAn_i$: Moving average of period $n$ at period $i$.
- $n$: Period.
- $C_i$: Closing price at period $i$.

$MA10_{i} = \frac{C_{i-9} + C_{i-8} + \cdots + C_{i-1} + C_i}{n}$

```python
fb = pd.read_csv("ticker-fb.csv", index_col=0, parse_dates=True)
```

```python
fb['Close'].rolling(2).sum().head()
```

```python
fb['Close'].rolling(3).mean().head()
```

```python
fb['MA50'] = fb['Close'].rolling(50).mean()
```

```python
fb.loc['2020-01-01':][['Close', 'MA50']].plot()
```

```python
fb['MA10'] = fb['Close'].rolling(10).mean()
```

```python
fb.loc['2020-10-01':][['Close', 'MA10']].plot()
```

```python
fb['Signal'] = fb['MA10'] < fb['Close']
```

```python
fb[['Close', 'MA10', 'Signal']].tail()
```
