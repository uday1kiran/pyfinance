```python
import pandas as pd
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
data = pd.read_csv("result-full.csv", index_col=0, header=[0, 1])
```

```python
data.head()
```

```python
price = pd.read_csv("result-price.csv", index_col=0)
```

```python
price.columns = ['Price']
```

```python
full_data = pd.concat([data, price], axis=1)
```

```python
full_data.head()
```

```python
focus_data = full_data[full_data['DE', 'D/E ratio'] & full_data['Current', 'Current'] & full_data['ROI', 'Mean']].copy()
```

```python
focus_data = focus_data[focus_data['BVPS', '1Y CAGR'] > 0.1].copy()
```

```python
focus_data.sort_values([('BVPS', '1Y CAGR')], ascending=False)
```

```python
focus_data.to_csv("result-focus.csv")
```
