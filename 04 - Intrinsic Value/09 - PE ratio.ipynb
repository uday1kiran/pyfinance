```python
import pandas as pd
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
pe = pd.read_csv("50-pe-ratio.csv", index_col=0)
```

```python
pe.tail()
```

```python
pe.index = pd.to_datetime(pe.index)
```

```python
pe.sort_index(ascending=False, inplace=True)
```

```python
pe.head()
```

```python
data = pe.resample('Q', convention='end').agg('mean')
```

```python
data = data.loc['2010-01-01':]
```

```python
data
```

```python
data = data.transpose()
```

```python
data['Mean'] = data.mean(axis=1)
```

```python
data.head()
```

```python
data['Mean'].to_csv("result-pe-avg.csv")
```
