```python
import pandas as pd
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
roi = pd.read_csv("50-roi.csv", index_col=0)
```

```python
roi.head()
```

```python
roi.tail()
```

```python
roi.index = pd.to_datetime(roi.index)
```

```python
roi.index
```

```python
roi.sort_index(ascending=False, inplace=True)
```

```python
roi.head()
```

```python
roi = roi.loc[:'2010-01-01']
```

```python
roi.tail()
```

```python
roi_t = roi.resample('Q', convention='end').agg('mean').transpose()
```

```python
roi_t['Mean'] = roi_t.apply(pd.Series.mean, axis=1)
```

```python
roi_t
```

```python
col = roi_t['Mean'] >= 10
```

```python
roi_t.sort_values('Mean', ascending=False)
```

```python
col.to_csv("result-roi.csv")
```
