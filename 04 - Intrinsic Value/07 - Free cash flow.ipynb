```python
import pandas as pd
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
fcf = pd.read_csv("50-fcf.csv", index_col=0)
```

```python
fcf.head()
```

```python
fcf = fcf.transpose()
```

```python
fcf.head()
```

```python
fcf['10Y CAGR'] = (fcf[2019]/fcf[2009])**(1/10) - 1
```

```python
fcf['5Y CAGR'] = (fcf[2019]/fcf[2014])**(1/5) - 1
```

```python
fcf['1Y CAGR'] = (fcf[2019]/fcf[2018])**(1/1) - 1
```

```python
fcf.head()
```

```python
data = fcf[['10Y CAGR', '5Y CAGR', '1Y CAGR']]
```

```python
data.sort_values('1Y CAGR', ascending=False)
```

```python
data.to_csv("result-fcf.csv")
```
