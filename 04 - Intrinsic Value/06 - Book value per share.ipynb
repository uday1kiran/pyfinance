```python
import pandas as pd
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
bvps = pd.read_csv("50-bvps.csv", index_col=0)
```

```python
bvps.tail()
```

```python
bvps.index = pd.to_datetime(bvps.index)
```

```python
bvps.index
```

```python
bvps.sort_index(ascending=False, inplace=True)
```

```python
bvps.head()
```

```python
bvps = bvps.resample('Q', convention='end').agg('mean')
```

```python
bvps = bvps.loc['2010-09-01':]
```

```python
bvps = bvps.transpose()
```

```python
bvps.head()
```

```python
result = pd.DataFrame(data=None, index=bvps.index)
```

```python
result['10Y CAGR'] = (bvps['2020-09-30']/bvps['2010-09-30'])**(1/10) - 1
```

```python
result['5Y CAGR'] = (bvps['2020-09-30']/bvps['2015-09-30'])**(1/5) - 1
```

```python
result['1Y CAGR'] = (bvps['2020-09-30']/bvps['2019-09-30'])**(1/1) - 1
```

```python
result.sort_values('1Y CAGR', ascending=False)
```

```python
result.to_csv("result-bvps.csv")
```
