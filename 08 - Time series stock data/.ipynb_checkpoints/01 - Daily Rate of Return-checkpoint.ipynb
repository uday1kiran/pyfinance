### Rate of return

\begin{equation*}
R = \frac{V_f - V_{init}}{V_{init}} = \frac{V_f}{V_{init}} - 1
\end{equation*}
- $R$: Simple rate of return
- $V_{init}$: Initial value
- $V_f$: Final value

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
stock_a = pd.read_csv("ticker-fb.csv", index_col=0, parse_dates=True)
```

```python
stock_a.shift()
```

```python
stock_a['RoR'] = stock_a['Adj Close']/stock_a['Adj Close'].shift(1) - 1
```

### %-change

\begin{equation*}
P_i = \frac{V_i}{V_{i-i}} - 1
\end{equation*}
- $P_i$: Pecentage change value
- $V_i$: Current value
- $V_{i-1}$: Previous value

```python
stock_a['RoR (v2)'] = stock_a['Adj Close'].pct_change()
```

### Normalization

\begin{equation*}
N_i = \frac{V_i}{V_{init}}*100
\end{equation*}
- $N_i$: Normalized value
- $V_{init}$: Initial value
