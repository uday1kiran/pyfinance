```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
tickers = ["FB", "NVDA", "AMZN", "HP", "AAPL"]

df_list = []
for ticker in tickers:
    df = pd.read_csv("ticker-" + ticker.lower() + ".csv", index_col=0, parse_dates=True)
    df_list.append(df['Adj Close'])

data = pd.concat(df_list, axis=1)
data.columns = tickers
```

```python
data = np.log(data/data.shift())
```

```python
data.corr()
```

```python
plt.imshow(data.corr(), interpolation='nearest')
```

```python
data.corr().style.background_gradient(cmap='Blues')
```
