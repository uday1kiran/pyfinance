```python
from yahoofinancials import YahooFinancials as YF
```

```python
tick = YF("AAPL")
```

```python
income_statement = tick.get_financial_stmts('annual', 'income', reformat=True)
```

```python
type(income_statement)
```

```python
aapl_is = income_statement['incomeStatementHistory']['AAPL']
```

```python
aapl_is
```

```python
import pandas as pd
```

```python
df = []

for d in aapl_is:
  df.append(pd.DataFrame.from_dict(d, orient='index'))

df_is = pd.concat(df)
```

```python
df_is
```
