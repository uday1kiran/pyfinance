### Step 1
- Import pandas datareader as pdr
- Import datetime as dt
- Import matplotlib.pyplot as plt
    - Remember %matplotlib notebook

```python
import pandas_datareader as pdr
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib notebook
```

### Step 2
- Make a datetime object start for 2020-01-01 (Jan. 1st, 2020)
- Make a datetime object end for 2020-12-31 (Dec. 31st, 2020)
- Read data from ticker NFLX from Yahoo! Finance with pandas datareader
- Take head on the dataframe

```python
start = dt.datetime(2020, 1, 1)
end = dt.datetime(2020, 12, 31)
```

```python
data = pdr.get_data_yahoo("NFLX", start, end)
```

```python
data.head()
```

### Step 3
- Make a figure with axis from matplotlib
- Make a plot of closing pring (Close)

```python
fig, ax = plt.subplots()
data['Close'].plot(ax=ax)
```

### Step 4
- Read data from 4 tickers: "NVDA", "HP", "FB", "GOOG"
- Use the same start and end date
- Also use Yahoo! Finance.
- Take head of the data

```python
tickers = ["NVDA", "HP", "FB", "GOOG"]
data = pdr.get_data_yahoo(tickers, start, end)
```

```python
data.head()
```

### Step 5
- Make 4 plots in 1 figure with the closing price (Close) of the 4 tickers from last step

```python
fig, ax = plt.subplots(2, 2)
data['Close', 'NVDA'].plot(ax=ax[0, 0], title='NVDA')
data['Close', 'HP'].plot(ax=ax[0, 1], title='HP')
data['Close', 'FB'].plot(ax=ax[1, 0], title='FB')
data['Close', 'GOOG'].plot(ax=ax[1, 1], title='GOOG')
plt.tight_layout()
```

### Step 6
- Create 4 DataFrames from the DataFrame from last step
- Each should contain the data for each ticker: "NVDA", "HP", "FB", "GOOG"

```python
nvda = data.xs('NVDA', axis=1, level=1)
hp = data.xs('HP', axis=1, level=1)
fb = data.xs('FB', axis=1, level=1)
goog = data.xs('GOOG', axis=1, level=1)
```

### Step 7
- Save the DataFrames from last step to CSV files
- Each DataFrame should be saved in their own CSV file

```python
nvda.to_csv("ticker-nvda.csv")
hp.to_csv("ticker-hp.csv")
fb.to_csv("ticker-fb.csv")
goog.to_csv("ticker-goog.csv")
```
