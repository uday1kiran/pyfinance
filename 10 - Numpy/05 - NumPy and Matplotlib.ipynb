```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
x = np.arange(1000)
y = 3*x + 300
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
x = np.arange(1000)
y = np.sqrt(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
x = np.random.random(5000)
y = np.random.random(5000)
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, alpha=0.3)
```

```python
x = np.random.randn(5000)
y = np.random.randn(5000)
```
